````python
# ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Îî•Îü¨ÎãùÏö©)
sequence_length = 30

def create_sequences(X, y, seq_len=30):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_len):
        X_seq.append(X[i:i+seq_len])
        y_seq.append(y[i+seq_len])
    return np.array(X_seq), np.array(y_seq)

X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train.values, sequence_length)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val.values, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test.values, sequence_length)

print(f"ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ shape: {X_train_seq.shape}")
print(f"  - ÏÉòÌîå Ïàò: {X_train_seq.shape[0]}")
print(f"  - ÏãúÌÄÄÏä§ Í∏∏Ïù¥: {X_train_seq.shape[1]}")
print(f"  - ÌäπÏÑ± Ïàò: {X_train_seq.shape[2]}")
````

## 2. ÏòàÏ†ú: LSTM Î™®Îç∏ + ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à Ï†ÑÎûµ

Ïù¥ ÏòàÏ†úÎäî 2Í∞ÄÏßÄ ÌïµÏã¨ ÏöîÏÜåÎ°ú Íµ¨ÏÑ±Îê©ÎãàÎã§:

1. **ÏòàÏ∏° Î™®Îç∏**: Í∞ÄÍ≤© ÏÉÅÏäπ/ÌïòÎùΩÏùÑ ÏòàÏ∏°ÌïòÍ≥† ÌôïÎ•† Ï†úÍ≥µ
2. **Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ**: ÏòàÏ∏° ÌôïÎ•†Ïóê Îî∞Îùº Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à

### Ï†ÑÎûµ ÏòàÏãú
- ÏÉÅÏäπ ÌôïÎ•† 90% ‚Üí ÏûêÎ≥∏Ïùò 90% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 60% ‚Üí ÏûêÎ≥∏Ïùò 60% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 40% ‚Üí Ìà¨ÏûêÌïòÏßÄ ÏïäÏùå (ÌòÑÍ∏à Î≥¥Ïú†)

````python
# ÏòàÏ†ú Î™®Îç∏: LSTM Í∏∞Î∞ò 2-class Î∂ÑÎ•òÍ∏∞
class ExampleLSTMModel(nn.Module):
    """
    ÏòàÏ†ú: LSTM Í∏∞Î∞ò Í∞ÄÍ≤© Î∞©Ìñ• ÏòàÏ∏° Î™®Îç∏
    Ï∂úÎ†•: ÏÉÅÏäπ ÌôïÎ•† (0~1)
    """
    def __init__(self, input_size, hidden_size=64, dropout=0.2):
        super(ExampleLSTMModel, self).__init__()
        
        self.lstm1 = nn.LSTM(input_size, hidden_size, batch_first=True, num_layers=1)
        self.dropout1 = nn.Dropout(dropout)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        
        self.lstm2 = nn.LSTM(hidden_size, hidden_size//2, batch_first=True, num_layers=1)
        self.dropout2 = nn.Dropout(dropout)
        self.bn2 = nn.BatchNorm1d(hidden_size//2)
        
        self.fc1 = nn.Linear(hidden_size//2, 16)
        self.relu = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout)
        
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        lstm_out, _ = self.lstm1(x)
        lstm_out = self.dropout1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        lstm_out = self.bn1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        
        lstm_out, _ = self.lstm2(lstm_out)
        lstm_out = self.dropout2(lstm_out[:, -1, :])
        lstm_out = self.bn2(lstm_out)
        
        out = self.fc1(lstm_out)
        out = self.relu(out)
        out = self.dropout3(out)
        
        out = self.fc2(out)
        out = self.sigmoid(out)
        
        return out

# ÏòàÏ†ú Î™®Îç∏ ÏÉùÏÑ±
example_model = ExampleLSTMModel(
    input_size=X_train_seq.shape[2],
    hidden_size=64,
    dropout=0.2
).to(device)

print("ÏòàÏ†ú Î™®Îç∏ Íµ¨Ï°∞:")
print(example_model)
print(f"\nTotal parameters: {sum(p.numel() for p in example_model.parameters()):,}")
````

````python
# ÌïôÏäµ Ìï®Ïàò (Ï†úÍ≥µÎê®)
def train_model(model, train_loader, val_loader, epochs=100, lr=0.001, patience=15):
    model = model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y.unsqueeze(1))
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            predicted = (outputs > 0.5).float()
            train_total += batch_y.size(0)
            train_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_train_loss = train_loss / len(train_loader)
        train_acc = train_correct / train_total
        
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs, batch_y.unsqueeze(1))
                
                val_loss += loss.item()
                predicted = (outputs > 0.5).float()
                val_total += batch_y.size(0)
                val_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_val_loss = val_loss / len(val_loader)
        val_acc = val_correct / val_total
        
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}')
        
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break
    
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    
    return history

print("‚úÖ ÌïôÏäµ Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

````python
# ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ
print("ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏãúÏûë...")
example_history = train_model(
    model=example_model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=100,
    lr=0.001,
    patience=15
)
print("\n‚úÖ ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å!")
````

ÏòàÏ∏° ÌôïÎ•†Ïóê Îî∞Îùº Ìà¨Ïûê ÎπÑÏú®ÏùÑ Ï°∞Ï†àÌïòÎäî Ï†ÑÎûµÏûÖÎãàÎã§.

````python
# ÏòàÏ∏° Ìï®Ïàò (ÌôïÎ•† Ìè¨Ìï®)
def predict_with_probability(model, data_loader):
    """
    Î™®Îç∏ ÏòàÏ∏° Î∞è ÌôïÎ•† Î∞òÌôò
    """
    model.eval()
    predictions_prob = []
    
    with torch.no_grad():
        for batch_X, _ in data_loader:
            batch_X = batch_X.to(device)
            outputs = model(batch_X)
            predictions_prob.append(outputs.cpu().numpy())
    
    predictions_prob = np.vstack(predictions_prob).flatten()
    predictions = (predictions_prob > 0.5).astype(int)
    
    return predictions_prob, predictions

# ÏòàÏ†ú Î™®Îç∏ ÏòàÏ∏°
example_prob, example_pred = predict_with_probability(example_model, test_loader)

print(f"ÏòàÏ∏° ÏôÑÎ£å!")
print(f"ÏòàÏ∏° Ïàò: {len(example_pred)}")
print(f"ÏÉÅÏäπ ÏòàÏ∏°: {np.sum(example_pred == 1)}Í∞ú")
print(f"ÌïòÎùΩ ÏòàÏ∏°: {np.sum(example_pred == 0)}Í∞ú")
print(f"\nÌèâÍ∑† ÏÉÅÏäπ ÌôïÎ•†: {example_prob.mean():.2%}")
````

````python
# ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ
def simulate_probability_trading(predictions_prob, actual_prices, dates, 
                                 initial_capital=10000, transaction_fee=0.001,
                                 threshold=0.5, position_scaling=True):
    """
    ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò
    
    Parameters:
    -----------
    predictions_prob : array
        ÏÉÅÏäπ ÌôïÎ•† (0~1)
    actual_prices : array
        Ïã§Ï†ú Í∞ÄÍ≤©
    dates : array
        ÎÇ†Ïßú
    initial_capital : float
        Ï¥àÍ∏∞ ÏûêÎ≥∏
    transaction_fee : float
        Í±∞Îûò ÏàòÏàòÎ£å
    threshold : float
        Îß§Ïàò Í≤∞Ï†ï ÏûÑÍ≥ÑÍ∞í (Í∏∞Î≥∏ 0.5)
    position_scaling : bool
        True: ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à
        False: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
    """
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    trade_log = []
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                sell_value = btc_holdings * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL_ALL',
                    'price': current_price,
                    'prob': prob,
                    'amount': btc_holdings,
                    'value': btc_holdings * current_price,
                    'fee': btc_holdings * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings = 0
            continue
        
        # Ìà¨Ïûê ÎπÑÏú® Í≤∞Ï†ï
        if position_scaling:
            # ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à
            if prob > threshold:
                # ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê (Ïòà: 70% ÌôïÎ•† ‚Üí 70% Ìà¨Ïûê)
                invest_ratio = prob
            else:
                # ÏûÑÍ≥ÑÍ∞í ÎØ∏ÎßåÏù¥Î©¥ Îß§ÎèÑ ÎòêÎäî Í¥ÄÎßù
                invest_ratio = 0
        else:
            # Îã®Ïàú Ï†ÑÎûµ: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
            invest_ratio = 1.0 if prob > threshold else 0.0
        
        # ÌòÑÏû¨ Ìè¨ÏßÄÏÖò ÎπÑÏú®
        current_position_ratio = (btc_holdings * current_price) / portfolio_value if portfolio_value > 0 else 0
        
        # Ìè¨ÏßÄÏÖò Ï°∞Ï†ï
        target_btc_value = portfolio_value * invest_ratio
        current_btc_value = btc_holdings * current_price
        
        if target_btc_value > current_btc_value:  # Îß§Ïàò ÌïÑÏöî
            buy_cash = target_btc_value - current_btc_value
            if buy_cash > cash:
                buy_cash = cash
            if buy_cash > 0:
                buy_amount = (buy_cash * (1 - transaction_fee)) / current_price
                btc_holdings += buy_amount
                trade_log.append({
                    'date': dates[i],
                    'action': 'BUY',
                    'price': current_price,
                    'prob': prob,
                    'amount': buy_amount,
                    'value': buy_cash,
                    'fee': buy_cash * transaction_fee
                })
                cash -= buy_cash
        elif target_btc_value < current_btc_value:  # Îß§ÎèÑ ÌïÑÏöî
            sell_btc = (current_btc_value - target_btc_value) / current_price
            if sell_btc > btc_holdings:
                sell_btc = btc_holdings
            if sell_btc > 0:
                sell_value = sell_btc * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL',
                    'price': current_price,
                    'prob': prob,
                    'amount': sell_btc,
                    'value': sell_btc * current_price,
                    'fee': sell_btc * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings -= sell_btc
    
    final_value = portfolio_values[-1]
    total_return = (final_value - initial_capital) / initial_capital * 100
    
    total_trade_volume = sum(trade['value'] for trade in trade_log)
    total_fees_paid = sum(trade['fee'] for trade in trade_log)
    
    return {
        'initial_capital': initial_capital,
        'final_value': final_value,
        'total_return': total_return,
        'portfolio_values': portfolio_values,
        'trade_log': trade_log,
        'num_trades': len(trade_log),
        'total_trade_volume': total_trade_volume,
        'total_fees_paid': total_fees_paid
    }

print("‚úÖ Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

```
   - **Í≥ºÏ†ÅÌï© Î∞©ÏßÄ (Dropout)**: 2-Layer Stacked LSTM Íµ¨Ï°∞ÏôÄ FC Î†àÏù¥Ïñ¥Ïóê Dropout(0.3)ÏùÑ Ï∂îÍ∞ÄÌïòÏó¨ Î™®Îç∏Ïù¥ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Ïóê Í≥ºÏ†ÅÌï©ÎêòÎäî Í≤ÉÏùÑ Î∞©ÏßÄÌïòÍ≥†, ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏùºÎ∞òÌôî ÏÑ±Îä•ÏùÑ ÌôïÎ≥¥ÌñàÏäµÎãàÎã§.

2. Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ: Ï∂îÏÑ∏ Ï∂îÏ¢Ö Í∏∞Î∞òÏùò ÌïòÏù¥Î∏åÎ¶¨Îìú AI Ï†ÑÎûµ (Ï∞ΩÏùòÏÑ±)
   Îã®ÏàúÌûà AIÏùò ÏòàÏ∏° Ïã†Ìò∏Îßå Îî∞Î•¥Îäî Í≤ÉÏù¥ ÏïÑÎãå, Í∏∞Ïà†Ï†Å Î∂ÑÏÑù ÏßÄÌëúÎ•º Í≤∞Ìï©ÌïòÏó¨ ÏãúÏû• ÏÉÅÌô©Ïóê Ïú†Ïó∞ÌïòÍ≤å ÎåÄÏùëÌïòÎäî ÌïòÏù¥Î∏åÎ¶¨Îìú ÏãúÏä§ÌÖúÏùÑ Íµ¨Ï∂ïÌñàÏäµÎãàÎã§.
   - **(ÌïµÏã¨) Ï∂îÏÑ∏ ÌïÑÌÑ∞ÎßÅ (MA 50)**: 'ÌïòÎùΩÏû•ÏóêÏÑúÎäî Îß§ÏàòÌïòÏßÄ ÏïäÎäîÎã§'Îäî Î≥¥ÏàòÏ†ÅÏù∏ Ìà¨Ïûê ÏõêÏπôÏùÑ ÏãúÏä§ÌÖúÌôîÌñàÏäµÎãàÎã§. ÌòÑÏû¨ Í∞ÄÍ≤©Ïù¥ 50Ïùº Ïù¥ÎèôÌèâÍ∑†ÏÑ† ÏúÑÏóê ÏûàÏùÑ ÎïåÎßå(ÏÉÅÏäπ Ï∂îÏÑ∏) Îß§Ïàò Ïã†Ìò∏Î•º Ïú†Ìö®ÌïòÍ≤å Ï≤òÎ¶¨ÌïòÏó¨, ÌÅ∞ ÌïòÎùΩÏû•ÏóêÏÑúÏùò ÏÜêÏã§ ÏúÑÌóòÏùÑ ÏõêÏ≤úÏ†ÅÏúºÎ°ú Ï∞®Îã®Ìï©ÎãàÎã§.
   - **ÏßÑÏûÖ Ï°∞Í±¥ (AI + RSI Î™®Î©òÌÖÄ)**: ÏÉÅÏäπ Ï∂îÏÑ∏ ÏÜçÏóêÏÑúÎèÑ RSI ÏßÄÌëúÍ∞Ä 70 Ïù¥ÏÉÅ(Í≥ºÎß§Ïàò)Ïùº Í≤ΩÏö∞, Ï∂îÍ≤© Îß§ÏàòÎ•º ÌîºÌïòÍ∏∞ ÏúÑÌï¥ ÏßÑÏûÖÏùÑ Î≥¥Î•òÌï©ÎãàÎã§. AIÏùò ÏÉÅÏäπ ÏòàÏ∏°Í≥º Í≥ºÏó¥ÎêòÏßÄ ÏïäÏùÄ ÏãúÏû• ÏÉÅÌô©Ïù¥ÎùºÎäî Îëê Í∞ÄÏßÄ Ï°∞Í±¥Ïù¥ Ï∂©Ï°±Îê† ÎïåÎßå ÏßÑÏûÖÌïòÏó¨ ÏäπÎ•†ÏùÑ ÎÜíÏûÖÎãàÎã§.
   - **Í≥µÍ≤©Ï†Å ÏßÑÏûÖ (Buy the Dip)**: ÏÉÅÏäπ Ï∂îÏÑ∏ Ï§ë ÏùºÏãúÏ†ÅÏù∏ Ï°∞Ï†ï(RSI 40 Ïù¥Ìïò)Ïù¥ Î∞úÏÉùÌïòÎ©¥, Ïù¥Î•º 'ÎàåÎ¶ºÎ™©' Îß§Ïàò Í∏∞ÌöåÎ°ú ÌåêÎã®ÌïòÍ≥† Î™®Îç∏Ïùò ÏòàÏ∏° ÌôïÎ•†Ïù¥ Îã§ÏÜå ÎÇÆÎçîÎùºÎèÑ Ï†ÅÍ∑πÏ†ÅÏúºÎ°ú Îß§Ïàò ÎπÑÏ§ëÏùÑ ÎäòÎ†§ ÏàòÏùµÏùÑ Í∑πÎåÄÌôîÌï©ÎãàÎã§.
   - **Î¶¨Ïä§ÌÅ¨ Í¥ÄÎ¶¨ (Exit Logic)**: Î™®Îç∏Ïù¥ ÌïòÎùΩÏùÑ Í∞ïÌïòÍ≤å ÏòàÏ∏°ÌïòÍ±∞ÎÇò(prob < 0.4), Í±∞ÏãúÏ†ÅÏù∏ Ï∂îÏÑ∏Í∞Ä Î∂ïÍ¥¥(Price < MA50)ÎêòÎ©¥ Ï¶âÏãú Îß§ÎèÑÌïòÏó¨ ÏÜêÏã§ÏùÑ Ï†úÌïúÌï©ÎãàÎã§.

3. ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï Í∑ºÍ±∞:
   - `sequence_length=30`: ÏïΩ Ìïú Îã¨Í∞ÑÏùò ÏãúÏû• Îç∞Ïù¥ÌÑ∞Î•º ÌïôÏäµÌïòÏó¨ Îã®Í∏∞ Î≥ÄÎèôÏÑ±Î≥¥Îã§Îäî Ï§ëÍ∏∞Ï†ÅÏù∏ Ìå®ÌÑ¥ÏùÑ Ìè¨Ï∞©ÌïòÎèÑÎ°ù ÏÑ§Ï†ïÌñàÏäµÎãàÎã§.
   - `hidden_size=64`: Î™®Îç∏Ïùò Î≥µÏû°ÎèÑÏôÄ Í≥ºÏ†ÅÌï© ÏúÑÌóò ÏÇ¨Ïù¥Ïùò Í∑†ÌòïÏùÑ Í≥†Î†§Ìïú ÌÅ¨Í∏∞ÏûÖÎãàÎã§.
   - `prob_threshold=0.55`: Î™®Îç∏Ïù¥ 55% Ïù¥ÏÉÅÏùò ÌôïÏã†ÏùÑ Í∞ÄÏßà ÎïåÎßå ÏßÑÏûÖÌïòÏó¨, Î∂àÌôïÏã§Ìïú Ïã†Ìò∏Î•º ÌïÑÌÑ∞ÎßÅÌï©ÎãàÎã§.
   - `MA_window=50`: Í∏àÏúµ ÏãúÏû•ÏóêÏÑú Ï§ëÍ∏∞ Ï∂îÏÑ∏Î•º ÌåêÎã®ÌïòÎäî Îç∞ Í∞ÄÏû• ÎÑêÎ¶¨ ÏÇ¨Ïö©ÎêòÎäî Í∏∞Ï§Ä Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.

4. ÏΩîÎìú ÌíàÏßà Î∞è Í∞ÄÎèÖÏÑ±:
   - **Î™®ÎìàÌôî**: ÌïµÏã¨ Î°úÏßÅÏùÑ `MyTradingModel` ÌÅ¥ÎûòÏä§ÏôÄ `simulate_my_strategy` Ìï®ÏàòÎ°ú Î∂ÑÎ¶¨ÌïòÏó¨ Ïû¨ÏÇ¨Ïö©ÏÑ±Í≥º Í∞ÄÎèÖÏÑ±ÏùÑ ÎÜíÏòÄÏäµÎãàÎã§.
   - **ÏÑ§Î™ÖÏ†ÅÏù∏ Î≥ÄÏàòÎ™Ö**: `prob_threshold`, `rsi_overbought` Îì± Î≥ÄÏàòÎ™ÖÏùÑ ÌÜµÌï¥ ÏΩîÎìúÏùò Ïó≠Ìï†ÏùÑ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ïù¥Ìï¥Ìï† Ïàò ÏûàÎèÑÎ°ù ÏûëÏÑ±ÌñàÏäµÎãàÎã§.
   - **Ï£ºÏÑù Î∞è Î¨∏ÏÑúÌôî**: Î≥µÏû°Ìïú Ï†ÑÎûµ Î°úÏßÅ Î∂ÄÎ∂ÑÏóê Îã®Í≥ÑÎ≥ÑÎ°ú Ï£ºÏÑùÏùÑ Ï∂îÍ∞ÄÌïòÏó¨ ÏΩîÎìúÏùò ÌùêÎ¶ÑÏùÑ ÏâΩÍ≤å Îî∞ÎùºÍ∞à Ïàò ÏûàÎèÑÎ°ù ÌñàÏúºÎ©∞, Ìï®ÏàòÏóêÎäî DocstringÏùÑ ÏûëÏÑ±ÌïòÏó¨ ÌååÎùºÎØ∏ÌÑ∞ÏôÄ Î∞òÌôò Í∞íÏùò ÏùòÎØ∏Î•º Î™ÖÌôïÌûà ÌñàÏäµÎãàÎã§.

5. ÏòàÏ†úÏôÄÏùò Ï∞®Î≥ÑÏ†ê:
   - **ÏãúÏä§ÌÖúÏ†Å Î¶¨Ïä§ÌÅ¨ Í¥ÄÎ¶¨**: ÏòàÏ†ú Î™®Îç∏Ïù¥ AI ÏòàÏ∏° ÌôïÎ•†ÏóêÎßå ÏùòÏ°¥ÌïòÎäî Î∞òÎ©¥, Î≥∏ Ï†ÑÎûµÏùÄ **ÏãúÏû•Ïùò Í±∞ÏãúÏ†ÅÏù∏ Ï∂îÏÑ∏(MA50)Î•º Î®ºÏ†Ä ÌåêÎã®**ÌïòÎäî 'ÏïàÏ†Ñ ÌïÑÌÑ∞'Î•º ÎèÑÏûÖÌñàÏäµÎãàÎã§. Ïù¥Î•º ÌÜµÌï¥ AI Î™®Îç∏Ïù¥ Îã®Í∏∞ ÎÖ∏Ïù¥Ï¶àÏóê Î∞òÏùëÌïòÏó¨ ÌïòÎùΩÏû•ÏóêÏÑú ÏûòÎ™ªÎêú Îß§Ïàò Ïã†Ìò∏Î•º ÎÇ¥Îäî ÏπòÎ™ÖÏ†ÅÏù∏ Ïã§ÏàòÎ•º ÏãúÏä§ÌÖúÏ†ÅÏúºÎ°ú Î∞©ÏßÄÌï©ÎãàÎã§.
   - **ÏÉÅÌô©Ïù∏ÏßÄÌòï Ìà¨Ïûê**: ÏãúÏû•Ïùò Í≥ºÏó¥ ÏÉÅÌÉú(RSI)ÏôÄ Ï∂îÏÑ∏ ÎÇ¥ Ï°∞Ï†ï Íµ≠Î©¥ÏùÑ Ïù∏ÏßÄÌïòÏó¨ Ìà¨Ïûê ÎπÑÏ§ëÏùÑ ÎèôÏ†ÅÏúºÎ°ú Ï°∞Ï†àÌïòÎäî ÎÖºÎ¶¨Î•º Ï∂îÍ∞ÄÌï®ÏúºÎ°úÏç®, Îã®Ïàú AI Ïã†Ìò∏ Í∏∞Î∞ò Îß§Îß§Î≥¥Îã§ Ìïú Îã®Í≥Ñ Î∞úÏ†ÑÎêú ÏùòÏÇ¨Í≤∞Ï†ï ÏãúÏä§ÌÖúÏùÑ Íµ¨ÌòÑÌñàÏäµÎãàÎã§.
```

````python
# TODO: ÏûêÏã†ÎßåÏùò Î™®Îç∏ Íµ¨ÌòÑ (LSTM + BatchNorm)

class MyTradingModel(nn.Module):
    """
    Stacked LSTMÍ≥º BatchNormÏùÑ ÌôúÏö©Ìïú ÏïàÏ†ïÏ†ÅÏù∏ Î™®Îç∏
    
    ÏûÖÎ†•: (batch, sequence_length, features)
    Ï∂úÎ†•: (batch, 1) - ÏÉÅÏäπ ÌôïÎ•† (0~1)
    """
    def __init__(self, input_size, hidden_size=64, num_layers=2, dropout=0.3):
        super(MyTradingModel, self).__init__()
        self.lstm = nn.LSTM(
            input_size, 
            hidden_size, 
            num_layers=num_layers, 
            batch_first=True, 
            dropout=dropout if num_layers > 1 else 0
        )
        self.bn = nn.BatchNorm1d(hidden_size)
        self.dropout1 = nn.Dropout(dropout)
        
        self.fc1 = nn.Linear(hidden_size, 32)
        self.relu = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout)
        
        self.fc2 = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        # LSTM layer
        lstm_out, _ = self.lstm(x)
        # ÎßàÏßÄÎßâ time stepÏùò Ï∂úÎ†•ÏùÑ ÏÇ¨Ïö©
        out = lstm_out[:, -1, :]
        
        # BatchNorm
        out = self.bn(out)
        out = self.dropout1(out)
        
        # Fully connected layers
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout2(out)
        
        out = self.fc2(out)
        out = self.sigmoid(out)
        
        return out

# Î™®Îç∏ ÏÉùÏÑ±
my_model = MyTradingModel(
    input_size=X_train_seq.shape[2],
    hidden_size=64,
    num_layers=2,
    dropout=0.3
).to(device)

print("ÎÇòÏùò Î™®Îç∏ Íµ¨Ï°∞:")
print(my_model)
print(f"\nTotal parameters: {sum(p.numel() for p in my_model.parameters()):,}")
````

````python
# Ï†ÑÎûµÏùÑ ÏúÑÌïú RSI Î∞è Ïù¥ÎèôÌèâÍ∑†(MA) Í∞í Ï§ÄÎπÑ
rsi_values = btc_features['RSI_14'].iloc[test_start_idx:test_start_idx+len(y_test_seq)].values
ma50_values = btc_features['MA_50'].iloc[test_start_idx:test_start_idx+len(y_test_seq)].values

print(f"Î≥¥Ï°∞ ÏßÄÌëú Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ ÏôÑÎ£å (RSI, MA50): {len(rsi_values)}Í∞ú")
````

````python
# TODO: ÏûêÏã†ÎßåÏùò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ Íµ¨ÌòÑ (Í≥µÍ≤©Ìòï ÏàòÏùµ Ï∂îÍµ¨ Ï†ÑÎûµ)
def simulate_my_strategy(predictions_prob, rsi_values, ma50_values, actual_prices, dates, 
                         initial_capital=10000, transaction_fee=0.001,
                         prob_threshold=0.5, rsi_overbought=80, rsi_oversold=30):
    """
    Í≥µÍ≤©Ìòï Ï†ÑÎûµ: Î™®Îç∏ ÏòàÏ∏° Ïã†Î¢∞ + ÌíÄÎß§Ïàò + RSI Í≥ºÏó¥ÏãúÎßå Îß§ÎèÑ
    """
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    trade_log = []
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        rsi = rsi_values[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                sell_value = btc_holdings * current_price
                cash += sell_value * (1 - transaction_fee)
                btc_holdings = 0
            continue
        
        # 1. Î™©Ìëú Ìà¨Ïûê ÎπÑÏú® Í≤∞Ï†ï (Í∏∞Î≥∏: ÌòÑÍ∏à Î≥¥Ïú†)
        invest_ratio = 0.0
        
        # 2. Îß§Ïàò Î°úÏßÅ (Í≥µÍ≤©Ï†Å)
        # Î™®Îç∏Ïù¥ ÏÉÅÏäπÏùÑ ÏòàÏ∏°ÌïòÍ±∞ÎÇò(>0.5), RSIÍ∞Ä Í≥ºÎß§ÎèÑ(<30) ÏÉÅÌÉúÎ©¥ Îß§Ïàò
        # Ï∂îÏÑ∏ ÌïÑÌÑ∞(MA50) Ï†úÍ±∞ÌïòÏó¨ Í±∞Îûò Í∏∞Ìöå ÌôïÎåÄ
        if prob >= prob_threshold or rsi < rsi_oversold:
            invest_ratio = 1.0  # Í≥µÍ≤©Ï†ÅÏúºÎ°ú 100% Ìà¨Ïûê
        
        # Îß§ÎèÑ Ï°∞Í±¥ (Í≥ºÎß§Ïàò Íµ¨Í∞ÑÏóêÏÑúÎäî Ï†ÑÎüâ Îß§ÎèÑÌïòÏó¨ Ïù¥Ïùµ Ïã§ÌòÑ)
        if rsi > rsi_overbought:
            invest_ratio = 0.0
        
        # 3. Ìè¨Ìä∏Ìè¥Î¶¨Ïò§ Î¶¨Î∞∏Îü∞Ïã± (Í∏∞Ï°¥ Ìï®Ïàò Î°úÏßÅÍ≥º Ïú†ÏÇ¨)
        target_btc_value = portfolio_value * invest_ratio
        current_btc_value = btc_holdings * current_price
        trade_amount = (target_btc_value - current_btc_value) / current_price
        
        if trade_amount > 0: # Îß§Ïàò
            buy_cash = trade_amount * current_price
            if buy_cash > cash: buy_cash = cash
            
            if buy_cash > 0:
                btc_holdings += (buy_cash * (1 - transaction_fee)) / current_price
                cash -= buy_cash
                trade_log.append({'date': dates[i], 'action': 'BUY', 'price': current_price, 'amount': buy_cash})
        
        elif trade_amount < 0: # Îß§ÎèÑ
            sell_amount = -trade_amount
            if sell_amount > btc_holdings: sell_amount = btc_holdings
            
            if sell_amount > 0:
                sell_value = sell_amount * current_price
                cash += sell_value * (1 - transaction_fee)
                btc_holdings -= sell_amount
                trade_log.append({'date': dates[i], 'action': 'SELL', 'price': current_price, 'amount': sell_value})
    
    # ÏµúÏ¢Ö Í≤∞Í≥º Í≥ÑÏÇ∞
    final_value = cash + btc_holdings * actual_prices[-1]
    total_return = (final_value - initial_capital) / initial_capital * 100
    total_fees_paid = sum(trade['amount'] * transaction_fee for trade in trade_log)
    
    return {
        'initial_capital': initial_capital, 'final_value': final_value,
        'total_return': total_return, 'portfolio_values': portfolio_values,
        'num_trades': len(trade_log), 'total_fees_paid': total_fees_paid
    }

    print("‚úÖ Í≥µÍ≤©Ï†ÅÏù∏ Ï†ÑÎûµ Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

````python
# ÏòàÏ∏° Î∞è Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò (ÎÇòÎßåÏùò Ï†ÑÎûµ)

# ÏòàÏ∏°
my_prob, my_pred = predict_with_probability(my_model, test_loader)

# Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò
my_result = simulate_my_strategy(
    predictions_prob=my_prob,
    rsi_values=rsi_values,
    ma50_values=ma50_values,
    actual_prices=test_prices_aligned,
    dates=test_dates_aligned,
    initial_capital=10000,
    transaction_fee=0.001,
    prob_threshold=0.5,
    rsi_overbought=80,
    rsi_oversold=30
)

print("="*70)
print("ÎÇòÏùò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ Í≤∞Í≥º")
print("="*70)
print(f"Ï¥àÍ∏∞ ÏûêÎ≥∏: ${my_result['initial_capital']:,.2f}")
print(f"ÏµúÏ¢Ö ÏûêÎ≥∏: ${my_result['final_value']:,.2f}")
print(f"ÏàòÏùµÎ•†: {my_result['total_return']:.2f}%")
print(f"Buy and Hold ÎåÄÎπÑ: {my_result['total_return'] - buy_hold_return:.2f}%p")
print(f"Í±∞Îûò ÌöüÏàò: {my_result['num_trades']}Ìöå")
print(f"Ï¥ù ÏàòÏàòÎ£å: ${my_result['total_fees_paid']:,.2f}")
print("="*70)
````
